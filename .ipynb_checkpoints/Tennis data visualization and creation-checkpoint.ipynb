```python
unique_locations = np.unique([data[i][1] for i in range(len(data))])
court_types = np.unique([data[i][5] for i in range(len(data))])
surface_types = np.unique([data[i][6] for i in range(len(data))])
tournament_round_types = np.unique([data[i][7] for i in range(len(data))])

print("{} unique locations".format(len(unique_locations)))
print("{} unique court types".format(len(court_types)))
print("{} unique surface types".format(len(surface_types)))
print("{} unique tournament round types".format(len(tournament_round_types)))
print("{} unique players".format(len(players)))
```

```python
def process_player_data(games, name):
    # First sort the games:
    #games.sort(key=lambda game:game[3])
    
    x = []
    y = []
    
    for i, game in enumerate(games):
        p1_winner = game[9] == name
    
        location = np.float(one_hot(game[1], len(unique_locations), split="+")[0])
        #date = game[3]
        court = np.float(one_hot(game[5], len(court_types), split="+")[0])
        surface = np.float(one_hot(game[6], len(surface_types), split="+")[0])
        tournament_round = np.float(one_hot(game[7], len(tournament_round_types), split="+")[0])
        
        p1 = game[9] if p1_winner else game[10]
        p1 = np.float(one_hot(p1, len(players), split="+")[0])
        
        p2 = game[10] if p1_winner else game[9]
        p2 = np.float(one_hot(p2, len(players), split="+")[0])
        
        p1_rank = game[11] if p1_winner else game[12]
        if p1_rank == "N/A":
            continue
        
        p2_rank = game[12] if p1_winner else game[11]
        if p2_rank == "N/A":
            continue
            
        p1_bet = game[28] if p1_winner else game[29]
        p1_bet = p1_bet if (p1_bet != "") else 0.0
        p2_bet = game[29] if p1_winner else game[28]
        p2_bet = p2_bet if (p2_bet != "") else 0.0
        
        x.append([location, court, surface, tournament_round
                          , p1, p2, p1_rank, p2_rank])
        if p1_winner: y.append(0.0) 
        else: y.append(1.0)
        
    return x, y
    
```

```python
for i in range(len(player_games)):
    bet = player_games[i][1][0][29]
    try:
        float(bet)
    except ValueError:
        print("Error with: ", bet)
```

```python
x_data, y_data = [], []
for player in player_games:
    px, py = process_player_data(player[1], player[0])
    x_data += px
    y_data += py

print(len(x_data))
for a, b in zip(x_data[0:5], y_data[0:5]):
    print(b, " ",a)
```

```python
input_dim = np.array(x_data[0]).shape
print(input_dim)
```

```python
model = Sequential()
model.add(Dense(64, activation="relu", input_dim=8))
model.add(Dense(64, activation="relu"))
model.add(Dense(64, activation="relu"))
model.add(Dropout(0.2))
model.add(Dense(64, activation="relu"))
model.add(Dense(1, activation="sigmoid"))

model.compile(optimizer="rmsprop", loss="binary_crossentropy", metrics=["accuracy"])
```

```python
model.fit(x_data, y_data, batch_size=64, validation_split=0.2, epochs=100, shuffle=True)
```
